```python
%matplotlib inline
import requests, json
import pandas as pd
```

First get all player data. Detailed player information doesn't contain the player name (it only contains player ID). Thus, we need to download the summary information for all player so that we can tie a player's ID to the name.

```python
PLAYER_DATA_ALL_URL = "https://fantasy.premierleague.com/drf/bootstrap-static"
all_data = requests.get(PLAYER_DATA_ALL_URL).json()
```

Here's an example of the player name and id

```python
first_name = all_data['elements'][0]['first_name']
second_name = all_data['elements'][0]['second_name']
player_id = all_data['elements'][0]['id']
print(first_name, second_name, player_id)
```

```python
first_name + ' ' + second_name
```

Now we need to loop through all the data and build a dictionary of player ID's and the corresponding names

```python
player_dict = {}

#The are less than 600 entries. to avoid 'IndexError: list index out of range' will skip
for i in range(600):
    try:
        first_name = all_data['elements'][i]['first_name']
        second_name = all_data['elements'][i]['second_name']
        player_name = first_name + ' ' + second_name
        player_id = all_data['elements'][i]['id']
        player_dict[i] = {'Player ID': player_id, 'Player Name': player_name} 
    except IndexError:
        continue
    except:
        print('Something went wrong getting player ids and names')
        raise
```

Now we have a list of player ID and the corresponding names we can fetch detailed information and associated it with the correct player
